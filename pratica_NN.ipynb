```python
import cv2
import numpy as np
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
```

```python
tree = ET.parse(r'.\assets\annotations\BikesHelmets0.xml')
```

```python
root = tree.getroot()

# Encontrar o elemento 'object' que cont√©m 'xmin'
for obj in root.findall('object'):
    # Acessar o elemento 'bndbox' dentro de 'object'
    bndbox = obj.find('bndbox')
    # Acessar o valor de 'xmin' dentro de 'bndbox'
    xmin = bndbox.find('xmin').text
    ymin = bndbox.find('ymin').text
    xmax= bndbox.find('xmax').text
    ymax= bndbox.find('xmax').text
    print('xmin:', xmin)
    print('ymin:', ymin)
    print('xmax:', xmax)
    print('ymax:', ymax)
    print('-'*20)

```

```python
img0= cv2.imread(r'assets\images\BikesHelmets0.png')
img0= cv2.cvtColor(img0,cv2.COLOR_BGR2RGB)
xmin, ymin, xmax, ymax = 29, 105, 82, 182
cv2.rectangle(img0, (xmin, ymin), (xmax, ymax), (0,255,0), 2)
cv2.rectangle(img0, (122, 66), (193, 193), (0,255,0), 2)

plt.imshow(img0)
```
